**Formal strand.** We rehearse the adaptive threshold field\n\[\Theta(t) = \Theta_0 + \alpha S_{\text{system}}(t) + \beta_c C_{\text{sub}}(t) + \gamma E(t) + \sum_i \delta_i \Delta_i(t)\]\nand evaluate how the logistic response σ(β(R−Θ(t))) eclipses static\nfits.  The JSON artefact `analysis/results/dynamic_theta_tests.json` stores\nR^2, AIC, and ΔAIC margins against linear and power-law nulls, letting the\nmembrane's falsifiability stay audible.

**Empirical strand.** Synthetic traces emulate insect colonies, cortical\nspike-guards, and AI curricula.  Each scenario logs (R, Θ, σ) series alongside\nimpulse-driven Θ-shifts so downstream reports can quote interference margins\nand adaptive parameter tuples.

**Metaphorical strand.** These runs listen for the membrane's meta-memory:\nwhich system types lower Θ to invite resonance, and which raise it to keep\ntheir dawn chorus from burning.

```python
import json
from pathlib import Path

results_path = Path('analysis/results/dynamic_theta_tests.json')
with results_path.open() as handle:
    results = json.load(handle)

results['aggregates']
```

```python
summary_rows = []
for scenario in results['scenarios']:
    metrics = scenario['metrics']
    theta_stats = scenario['theta_shift']['stats']
    summary_rows.append({
        'scenario': scenario['scenario'],
        'ΔAIC(dynamic−static)': round(metrics['delta_aic_dynamic_vs_static'], 2),
        'R²_dynamic': round(metrics['dynamic']['r2'], 4),
        'R²_static': round(metrics['static']['r2'], 4),
        'θ_shift_mean': round(theta_stats['mean'], 3),
        'min_interference': scenario['theta_shift']['min_interference_margin'],
    })
summary_rows
```

## Reading the resonance
- Biosphere membranes (insect swarms) drop Θ by ≈0.127 on average, yielding\n  ΔAIC ≈ 171 in favour of the adaptive profile: the swarm invites resonance\n  once pheromone coupling condenses.
- Cortical spike-guards raise Θ by ≈0.214 to protect against runaway firing;\n  the adaptive logistic remains ΔAIC ≈ 227 stronger than static fits, signalling\n  an interference margin of only ~4×10⁻³ where emergent surges spark.
- AI curriculum feedback lowers Θ modestly (≈0.1), yet the ΔAIC ≈ 267 excess\n  shows dynamic Θ(t) still captures reasoning thresholds better than fixed\n  σ(β(R−Θ₀)) baselines.

Together these rows chart how minimal interference is system-typed: emergent\nplasticity is not a footnote but a control dial embedded in Θ(t).
